Como deduzimos em sala de aula que

$$ W(n_1) = \frac{N!}{n_1!(N - n_1)!}p^{n_1}q^{N-n_1} $$

Podemos montar um gráfico disso facilmente em python e demonstrar o gráfico com $N=60$, $p=2/3$ e $q=1/3$:

```python
import matplotlib.pyplot as plt # pacote que produz o gráfico
import numpy as np # pacote que facilita as operações matemáticas com listas


#configuração de estilo da plotagem
plt.style.use('classic')
plt.style.use('dark_background')


#criando a função W(n1)
def W(n,p,q,N):
  n = np.array(n, dtype=int).astype(int)

  n_fact = np.array([np.math.factorial(n_i) for n_i in n])
  N_fact = np.math.factorial(N)
  N_n_fact = np.array([np.math.factorial(N - n[i]) for i in range(len(n))])

  A = N_fact/(n_fact*N_n_fact)
  B = np.power(p,n)*np.power(q,N-n)

  return A*B



#produção dos dados
N = 60
space = np.linspace(1,N,N)
p = 2/3
q = p - 1 #1/3

res = W(space,p,q,N)



# plotagem
plt.title("Random Walk\n"+r"$p = 2/3$ | $q = 1/3$ | $N = 60$",fontweight="bold")

plt.plot(space,abs(res),color='red',linewidth=1.5)

plt.annotate(r"$ W(n_1) = \frac{N!}{n_1!(N - n_1)!}p^{n_1}q^{N-n_1}$",xy=(5,0.08),size=20)

plt.xlabel(r"$n_i$ passos")
plt.ylabel(r"$W(n)$")
plt.grid(True, linestyle='--', linewidth=0.1)
plt.show()
```
