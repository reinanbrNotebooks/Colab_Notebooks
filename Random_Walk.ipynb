```python
import matplotlib.pyplot as plt
import matplotlib._color_data as mcd
import statistics as sts
import numpy as np

plt.style.use('seaborn')
```

```python
def random_walk(N,valv=.5):
  s = [[0],[0]]
  for _ in range(N):
    ds_x = np.random.uniform(-1,1)*valv
    ds_y = np.random.uniform(-1,1)*valv
    s_x = s[0][-1]+ds_x
    s_y = s[1][-1]+ds_y
    s[0].append(s_x)
    s[1].append(s_y)
  s_ = np.array(s)
  # s_way = np.array([np.sqrt(p[0]**2 + p[1]**2) for p in zip(s_[0],s_[1])])
  # s_way_sum = s_way.sum()
  # d = np.sqrt(s_[0][-1]**2 + s_[1][-1]**2)
  return s_
```

```python
N_steps = 1000
N_walks  = 50
steps = list(range(N_steps+1))
s_list = []
d_list = []
s_way_list = []
s_way_sum_list = []

fig, ax = plt.subplots(dpi=180,figsize=(12,19),nrows=2, ncols=1)
ax_list = []
for i in range(N_walks):
  s = random_walk(N_steps,valv=.1)
  d = np.sqrt(s[0][-1]**2 + s[1][-1]**2)
  s_way = np.array([np.sqrt(p[0]**2 + p[1]**2) for p in zip(s[0],s[1])])
  s_way_sum = s_way.sum()


  s_list.append(s)
  s_way_list.append(s_way)
  d_list.append(d)
  s_way_sum_list.append(s_way_sum)


  ax[0].plot(s[0],s[1])
  ax[1].plot(steps,s_way)

  #print('fazendo a plotagem:', i)
  #plt.annotate("ínicio",xy=(0,0))
  #plt.annotate('fim',xy=(s[0][-1],s[1][-1]))



way_std = sts.stdev(s_way_sum_list)
way_mean = sts.mean(s_way_sum_list)

ax[0].set_title(f'Todos Os {N_walks} Passeios Aleatórios Com {N_steps} Passos Cada\n[ Média = {way_mean:.3f} | Desvio Padrão = {way_std:.3f} ]')
ax[0].set_xlabel('x')
ax[0].set_ylabel('y')


d_std = sts.stdev(d_list)
d_mean = sts.mean(d_list)

ax[1].set_title(f"Todas As Distâncias Percorridas Em Cada Passeio Aleatório\n[Média = {d_mean:.3f} | Desvio Padrão = {d_std:.3f} ]")
ax[1].set_xlabel('passos')
ax[1].set_ylabel('Distância')

plt.savefig('random_walk.png',dpi=300)
plt.show()
```
